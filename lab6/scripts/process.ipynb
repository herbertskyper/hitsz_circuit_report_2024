```python
# ()
import numpy as np
import matplotlib.pyplot as plt
import random

# 
colors = ['g', 'r', 'c', 'm']
color_random = random.choice(colors)  # 


# 
def square_wave(A, T, t, duty_cycle):
    return A * (t % T < T * duty_cycle) * 2 - A

def triangle_wave(A, T, t):
    C=1e-7
    t[0]=100
    return np.where(t%(2*T)<=T, A * (1 - np.exp(-t%(T) / (C /T)))+3000000, A * (np.exp(-t%(T) / (C /T)))-3200000)/1000000

# 
f = 435.1  # 403.6Hz
A = 5.9  # 23.6V
duty_cycle = 0.5  # 50%

# 
T = 1 / f

# 
t = np.linspace(0, 3 * T, 512) # 3



duty_cycle1 = 0.5  # 47.37%
f2 = 839.9  # 1000Hz
A2 = 6.32 / 2  # 4V

# 
T2 = 1 / f2

plt.figure(figsize=(10, 6))
plt.title('Square Wave (Rf = 10k)')
# 
plt.text(0.05, 0.8, 'Frequency: {:.2f}Hz'.format(f), fontsize=12, color='black', transform=plt.gca().transAxes)
plt.text(0.05, 0.75, 'Amplitude: {:.2f}V'.format(A), fontsize=12, color='black', transform=plt.gca().transAxes)
# plt.text(0.05, 0.7, 'Duty Cycle: {:.2f}'.format(duty_cycle), fontsize=12, color='black', transform=plt.gca().transAxes)

plt.text(0.05, 0.95, 'Frequency: {:.2f}Hz'.format(f2), fontsize=12, color='black', transform=plt.gca().transAxes)
plt.text(0.05, 0.9, 'Amplitude: {:.2f}V'.format(A2), fontsize=12, color='black', transform=plt.gca().transAxes)
# plt.text(0.05, 0.7, 'Duty Cycle: {:.2f}'.format(duty_cycle), fontsize=12, color='black', transform=plt.gca().transAxes)

plt.plot(1000 * t, square_wave(A, T, t, duty_cycle), label='Square Wave', color=color_random)
plt.plot(1000 * t, triangle_wave(A2, T2, t), label='Capacitance Wave', color='orange')
plt.xlabel('Time / ms')
plt.ylabel('Voltage / V')
# y
yticks = np.arange(-int(A+2)*10/10, int(A+2)*10/10, 2)  # -30304
plt.yticks(yticks)
plt.ylim(-10,10)
plt.legend()  # 

# 
plt.savefig('Square Wave.png')
plt.show()


```

```python
#  
# 
# 
# 
# 
# 

# 

T1 = 1 / 1000  # 1ms,
A1 = 11.6 / 2  # 12.8V
duty_cycle1 = 0.5  # 47.37%
f2 = 1000  # 1000Hz
A2 = 4 / 2  # 4V

# 
T2 = 1 / f2

# 
t = np.linspace(0, 3 * T1, 512)  # 3


# 
def triangle_wave(A, T, t):
    return 4 * A / T * np.abs(t % T - T / 2) - A


# 
plt.figure(figsize=(10, 6))

plt.title('Square Wave and Triangle Wave')
# 
plt.text(0.1, 0.8, 'Duty Cycle: {:.2f}'.format(duty_cycle1), fontsize=12, color='black', transform=plt.gca().transAxes)
plt.text(0.1, 0.75, 'Square Wave Amplitude: {:.2f}V'.format(A1), fontsize=12, color='black',
         transform=plt.gca().transAxes)
plt.text(0.1, 0.7, 'Triangle Wave Amplitude: {:.2f}V'.format(A2), fontsize=12, color='black',
         transform=plt.gca().transAxes)
plt.text(0.1, 0.65, 'Triangle Wave Frequency: {:.2f}Hz'.format(f2), fontsize=12, color='black',
         transform=plt.gca().transAxes)

plt.plot(1000 * t, square_wave(A1, T1, t, duty_cycle1), label='Square Wave', color=color_random)
plt.plot(1000 * t, triangle_wave(A2, T2, t), label='Triangle Wave', color='orange')
plt.xlabel('Time / ms')
plt.ylabel('Voltage / V')
# y
yticks = np.arange(-8, 8.1, 1)  # -881
plt.yticks(yticks)
plt.grid()
plt.legend()
# 
plt.savefig('Square Wave and Triangle Wave.png')
plt.show()
```
