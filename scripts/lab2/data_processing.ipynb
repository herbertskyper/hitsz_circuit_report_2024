```python
# -*encoding:gbk
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data=pd.read_excel('data.xlsx')
plt.plot(data.loc[:,0],data.loc[:,'7.5V'])
plt.xlabel(r'$U_{G2K}$')
plt.ylabel(r'$I_A$',rotation=0)
plt.title(r'$I_A$-$U_{G2K}$ when $U_{G2A}=7.50V$')
plt.show()
```

```python
plt.plot(data.loc[:,0],data.loc[:,'8V'])
plt.xlabel(r'$U_{G2K}$')
plt.ylabel(r'$I_A$',rotation=0)
plt.title(r'$I_A$-$U_{G2K}$ when $U_{G2A}=8.00V$')
plt.show()
```

```python
# -*encoding:gbk
# numpymatplotlib
import numpy as np
import matplotlib.pyplot as plt



# xy
x = data.loc[:,0]
y = data.loc[:,'7.5V']

# polyfit

# 
plt.plot(x, y)

# 
#plt.plot(x_fit, y_fit, 'r', label='fit')

# # scipy
from scipy.signal import find_peaks

peaks, _ = find_peaks(y)
for peak in peaks:
    plt.axvline(x=x[peak], ymax=y[peak]/500,linestyle='--',color='r')

plt.ylim(0, 500)
plt.xticks(x[peaks])



# 
plt.show()
```
