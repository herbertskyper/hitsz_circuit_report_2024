```python
import matplotlib.pyplot as plt
import numpy as np

# 定义两个波的峰峰值，周期和相移
T=20
amplitude1, period1, phase_shift1 = 30, T, 0
amplitude2, period2, phase_shift2 = 32, T, -10.68

phase_shift1=np.radians(phase_shift1)
phase_shift2=np.radians(phase_shift2)

period1=2*np.pi/period1
period2=2*np.pi/period2

amplitude1/=2
amplitude2/=2

# 生成x轴的数据
x = np.linspace(0, 100, 100)

# 生成两个正弦波的数据
y1 = amplitude1 * np.sin(period1 * x + phase_shift1)
y2 = amplitude2 * np.sin(period2 * x + phase_shift2)
plt.xlabel("t(ms)")
plt.ylabel("U(V) I(A)")
# 绘制两个正弦波
plt.plot(x, y1, label='U')
plt.plot(x, y2, label='I')
plt.legend()
plt.show()
```
